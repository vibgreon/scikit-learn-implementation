## Linear Regression Model

```python
# after looking into plot, we see some trend in alcohol and proline

from sklearn.linear_model import LinearRegression
```

> Calling model

```python
# calling model

reg = LinearRegression()
reg
```

```python
# X should be a 2D array or pandas DataFrame
# y can be a 1D array or 1 dimentional DataFrame

# X = df['proline']  This is a pandas series
y = df['alcohol']

# So X is a pandas series rn, which is one dimensional
# X

# easily achieved like this
X = df[['proline']]
```

> Fitting Data

```python
# fitting data

reg.fit(X,y)
```

> train on X

```python
# prediction

pred = reg.predict(X)
```

> to check accuracy

```python
from sklearn import metrics
```

```python
# r2_score is use for regression cases (not for classifiers)
# compare  actual output with predicted output
# not satisfactory result, considering we are only taking one variable

metrics.r2_score(y, pred)
```

## Multiple Linear Regression model

```python
# dropping 'wine_type' because it is a categorical variable
# dropping 'alcohol' because that is our output variable

X = df.drop(['alcohol','wine_target'], axis=1)
```
