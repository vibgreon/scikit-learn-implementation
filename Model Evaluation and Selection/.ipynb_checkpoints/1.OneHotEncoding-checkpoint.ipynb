## Manipulate data

> change values

```python
# this will change values
# where num is 0 -> 0
# if num != 0 -> 1

df['num'] = np.where(df.num == 0, 0, 1)
```

```python
df.num.unique()
```

```python
# we can write both ways

df.ca.unique()
# df['ca'].unique()
```

```python
# '?' is representing the missing value

df.thal.unique()
```

> replce missing values to nan

```python
# we replace '?' to nan
# inplace=True will make the change effective in the data

df.replace('?',np.nan, inplace=True)
```

```python
df['ca'].unique()
```

```python
df.thal.unique()
```

```python
# df.ca = df.ca.astype(np.float)
df.ca = df.ca.astype(float)
```

## Visualizing Data

```python
g = sns.PairGrid(data=df, diag_sharey=False, hue='num')
# g.map_diag(plt.hist, bins=20, alpha=0.9, density=False, histtype='barstacked')
g.map_diag(sns.kdeplot)
g.map_upper(plt.scatter, alpha = 0.5)
g.map_lower(sns.regplot, fit_reg=False,scatter_kws={'alpha':0.3})

g.add_legend()
```

## Remove nan values

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='most_frequent')
```

> fit so that it learns to put most frequently happening values

```python
imputer.fit(X_train)
```

> transform data and save it back to original data

```python
X_train= imputer.transform(X_train)
X_test = imputer.transform(X_test)
```
